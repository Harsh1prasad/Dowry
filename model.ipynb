```python
from faker import Faker
import pandas as pd
import random
import numpy as np
```

```python
fake = Faker("en_IN")
```

```python
# possible type
body_type=["slim","athletic","avrage","heavy"]
job_type=["private job","government job","business","startup","self employee"]
family_occupation=["business","service","agriculture","politics","professional"]
religion=["hindu","muslim","sikh","christin","jain","buddhist"]
cast=["general","obc","sc","st"]
fairness=["very fair","fair","wheatish","bark"]
```

```python
def generate_boy():
    return{
        "name":fake.name_male(),
        "age":random.randint(21,48),
        "height_cm":random.randint(155,200),
        "body_type":random.choice(body_type),
        "job_type":random.choice(job_type),
        "salary_lpa":round(random.uniform(3,40),2),
        "family_networth":round(random.uniform(10,500),2),
        "family_occupation":random.choice(family_occupation),
        "religion":random.choice(religion),
        "cast":random.choice(cast),
        "nri":random.choice(["yes","no"]),
        "fairness":random.choice(fairness),
        "dowry":round(random.uniform(10,150),2)
        
    }
```

```python
data = [generate_boy() for _ in range(10000)]
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv("dowry.csv",index=False)
```

```python
import matplotlib.pyplot as pt
import seaborn as sns
```

```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder,LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
# body_type - ordinal encodeing
# job type - ordinal encoding
# framily_occupation - ordinal encoding
# religion - ordinal encoding
# cast - ordinal encoding
# nri - one hot encoding
# fairness - ordinal encoding
df.columns
```

```python
df.info()
```

```python
num_col = df.select_dtypes(include=["int","float"]).columns
```

```python
sns.boxplot(df[num_col[3]])
```

```python
transformer = ColumnTransformer(transformers = [
    ("t1",OrdinalEncoder(categories=[["athletic","avrage","slim","heavy"]]),["body_type"]),
    ("t2",OrdinalEncoder(categories=[["government job","business","startup","private job","self employee"]]),["job_type"]),
    ("t3",OrdinalEncoder(categories=[["politics","business","service","professional","agriculture"]]),["family_occupation"]),
    ("t4",OneHotEncoder(categories=[['buddhist', 'christin', 'jain', 'muslim', 'hindu', 'sikh']],sparse_output=True),["religion"]),
    ("t5",OneHotEncoder(categories=[['sc', 'general', 'st', 'obc']],sparse_output=True),["cast"]),
    ("t6",OneHotEncoder(categories=[["no","yes"]],sparse_output=True),["nri"]),
    ("t7",OrdinalEncoder(categories=[['wheatish', 'fair', 'bark', 'very fair']]),["fairness"]),
    ("t8",StandardScaler(),['age', 'height_cm', 'salary_lpa', 'family_networth','dowry']),
],remainder="passthrough",verbose_feature_names_out=False)
```

```python
df1 = transformer.fit_transform(df)
features = transformer.get_feature_names_out()
```

```python
df1.shape
```

```python
df2 = pd.DataFrame(data=df1,columns=features)
```

```python
df2.drop(columns=["name"],inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error , mean_absolute_error,r2_score
import matplotlib.pyplot as pt
from sklearn.tree import plot_tree
import time
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(df2.iloc[:,0:-1],df2.iloc[:,-1],test_size=0.2,random_state=42)
print(Xtrain.shape,
Ytrain.shape,
Xtest.shape,
Ytest.shape)
```

```python
# Random Forest Model
start = time.time()
rfr=RandomForestRegressor(n_estimators=100,criterion='squared_error',max_depth=10)
# model creation
start = time.time()
score = cross_val_score(rfr,Xtrain,Ytrain,cv=5,scoring="neg_mean_squared_error")
end = time.time()
print(f"time taken is {end-start}")
```

```python
print(f"score of the model is {-score.mean()}")
```

```python
rfr.fit(Xtrain,Ytrain)
```

```python
predict = rfr.predict(Xtest)
```

```python
print(f"mean squared error is {mean_squared_error(predict,Ytest)}")
```
